Label Encoding for 'Race', 'Gender', 'Change', 'diabetesMed'

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
features_to_label_encode = ['race', 'gender', 'change', 'diabetesMed']

for feature in features_to_label_encode:
    temp_diabetes[feature] = label_encoder.fit_transform(temp_diabetes[feature])
```

Ordinal Encoding for 'max_glu_serum', 'A1Cresult', and similar:
For 'max_glu_serum' and 'A1Cresult', There is clear ordinal data including a 'None' category. First map these categories to an ordinal scale. Assuming 'None' indicates not measured and should be considered as a separate category

```python
from sklearn.preprocessing import OrdinalEncoder

# Mapping
age_mapping = {'[0-10)':0, '[10-20)':1, '[20-30)':2, '[30-40)':3, '[40-50)':4, '[50-60)':5, '[60-70)':6, '[70-80)':7,'[80-90)':8, '[90-100)':9}
max_glu_serum_mapping = {'None': 0, 'Norm': 1, '>200': 2, '>300': 3}
A1Cresult_mapping = {'None': 0, 'Norm': 1, '>7': 2, '>8': 3}
metformin_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
repaglinide_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
nateglinide_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
chlorpropamide_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
glimepiride_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
acetohexamide_mapping = {'No':0, 'Steady': 1}
glipizide_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
glyburide_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
tolbutamide_mapping = {'No':0, 'Steady': 1}
pioglitazone_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
rosiglitazone_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
acarbose_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
miglitol_mapping = {'No':0, 'Steady': 1, 'Up':2, 'Down':3}
troglitazone_mapping = {'No':0, 'Steady': 1}
tolazamide_mapping = {'No':0, 'Steady': 1, 'Up':2}
examide_mapping = {'No':0}
citoglipton_mapping = {"No": 0}
insulin_mapping = {"No": 0, "Steady": 1, "Up": 2, "Down": 3}
glyburide_metformin_mapping = {"No": 0, "Steady": 1, "Up": 2, "Down": 3}
glipizide_metformin_mapping = {"No": 0, "Steady": 1}
glimepiride_pioglitazone_mapping = {"No": 0, "Steady": 1}
metformin_rosiglitazone_mapping = {'No':0}
metformin_pioglitazone_mapping = {'No':0, 'Steady': 1}
change_mapping = {'No':0, 'Ch':1}
diabetesMed_mapping = {'No':0, 'Yes':1}

# Applying mapping
temp_diabetes['age'] = temp_diabetes['age'].map(age_mapping)
temp_diabetes['max_glu_serum'] = temp_diabetes['max_glu_serum'].map(max_glu_serum_mapping)
temp_diabetes['A1Cresult'] = temp_diabetes['A1Cresult'].map(A1Cresult_mapping)
temp_diabetes['metformin'] = temp_diabetes['metformin'].map(metformin_mapping)
temp_diabetes['repaglinide'] = temp_diabetes['repaglinide'].map(repaglinide_mapping)
temp_diabetes['nateglinide'] = temp_diabetes['nateglinide'].map(nateglinide_mapping)
temp_diabetes['chlorpropamide'] = temp_diabetes['chlorpropamide'].map(chlorpropamide_mapping)
temp_diabetes['glimepiride'] = temp_diabetes['glimepiride'].map(glimepiride_mapping)
temp_diabetes['acetohexamide'] = temp_diabetes['acetohexamide'].map(acetohexamide_mapping)
temp_diabetes['glipizide'] = temp_diabetes['glipizide'].map(glipizide_mapping)
temp_diabetes['glyburide'] = temp_diabetes['glyburide'].map(glyburide_mapping)
temp_diabetes['tolbutamide'] = temp_diabetes['tolbutamide'].map(tolbutamide_mapping)
temp_diabetes['pioglitazone'] = temp_diabetes['pioglitazone'].map(pioglitazone_mapping)
temp_diabetes['rosiglitazone'] = temp_diabetes['rosiglitazone'].map(rosiglitazone_mapping)
temp_diabetes['acarbose'] = temp_diabetes['acarbose'].map(acarbose_mapping)
temp_diabetes['miglitol'] = temp_diabetes['miglitol'].map(miglitol_mapping)
temp_diabetes['troglitazone'] = temp_diabetes['troglitazone'].map(troglitazone_mapping)
temp_diabetes['tolazamide'] = temp_diabetes['tolazamide'].map(tolazamide_mapping)
temp_diabetes['examide'] = temp_diabetes['examide'].map(examide_mapping)
temp_diabetes['citoglipton'] = temp_diabetes['citoglipton'].map(citoglipton_mapping)
temp_diabetes['insulin'] = temp_diabetes['insulin'].map(insulin_mapping)
temp_diabetes['glyburide-metformin'] = temp_diabetes['glyburide-metformin'].map(glyburide_metformin_mapping)
temp_diabetes['glipizide-metformin'] = temp_diabetes['glipizide-metformin'].map(glipizide_metformin_mapping)
temp_diabetes['glimepiride-pioglitazone'] = temp_diabetes['glimepiride-pioglitazone'].map(glimepiride_pioglitazone_mapping)
temp_diabetes['metformin-rosiglitazone'] = temp_diabetes['metformin-rosiglitazone'].map(metformin_rosiglitazone_mapping)
temp_diabetes['metformin-pioglitazone'] = temp_diabetes['metformin-pioglitazone'].map(metformin_pioglitazone_mapping)
temp_diabetes['change'] = temp_diabetes['change'].map(change_mapping)
temp_diabetes['diabetesMed'] = temp_diabetes['diabetesMed'].map(diabetesMed_mapping)

```
